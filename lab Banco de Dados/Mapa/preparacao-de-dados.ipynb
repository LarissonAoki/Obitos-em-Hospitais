```python
import codecs
import pandas as pd
# Fontes: IBGE, Datasus
# Coordenadas: https://drive.google.com/open?id=1Chd8ZjPCZwpmtCcQG-2V4dSB_vuZcEAC
# Sepse 2018: https://drive.google.com/open?id=1L7n0e4EGCjvcAVhKKrkrdghrTsP1xivA
# Pneumonia 2018: https://drive.google.com/open?id=1RxClZFfN8a4VZmo4usYO71IrF2hrP--R

dfCoord = pd.read_csv('Coordenadas_Municipios_2010.csv', low_memory=False,delimiter=';', decimal=',')
dfSepse = pd.read_csv('Sepse por Municipio 2018.csv', low_memory=False,delimiter=';', decimal=',')
dfPneumo = pd.read_csv('Pneumo por Municipio 2018.csv', low_memory=False,delimiter=';', decimal=',')
```

```python
cidadesLista = ['municipios']
obitosLista = [0]
longLista = [0]
latLista = [0]

for _, municipio in dfSepse.iterrows():
    cidade = municipio['Municipios'].split(' ', 1)[1]
    cidade = cidade.upper()
    for _, coordenada in dfCoord.iterrows():
        if coordenada['NOME_MUNICIPIO'] in cidade:
                cid = municipio['Municipios'].split(' ', 1)[1]
                print(cid)
                cidadesLista.append(cid)
                obitosLista.append(municipio['Obitos'])
                longLista.append(coordenada['LONGITUDE'])
                latLista.append(coordenada['LATITUDE'])
                cidade = ''
                cid = ''
```

```python

f = codecs.open('sepseCoord.csv', 'w', "utf-8")
s = ""

f.write("Municipios;Obitos;Longitude;Latitude\n")

for c in range(len(cidadesLista)):
    s = cidadesLista[c] + ";" + str(obitosLista[c]) + ";" + str(longLista[c]) + ";" + str(latLista[c]) +"\n"
    f.write(s)

f.close()
```
