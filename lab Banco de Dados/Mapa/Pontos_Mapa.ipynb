```python
import pandas as pd
import folium
```

```python
dfSepse = pd.read_csv('sepseCoord.csv', low_memory=False,delimiter=';', decimal=',')
dfSepse.head(10)
```

```python
brasil = folium.Map(
    location=[-16.1237611, -59.9219642],
    zoom_start=4
)
```

```python
brasil_com_pontos_sepse = brasil

for _, Municipio in dfSepse.iterrows():
    if (int(Municipio['Obitos']) >= 400) :
            c='crimson',
    elif (int(Municipio['Obitos']) <= 100) :
            c='yellow',
    else :
            c='orange',
    folium.CircleMarker(
        location=[float(Municipio['Latitude']), float(Municipio['Longitude'])],
        radius=3,
        color=c,  
        fill=True,
        fill_opacity=1.0,
        tooltip=str(Municipio['Obitos']),
        stroke=False
).add_to(brasil_com_pontos_sepse)
```
