```python
import pandas as pd
import matplotlib.pyplot as plt

colors = ['#2678B2', '#FD7F28']
```

```python
df = pd.read_csv('doencas morte.csv')

colunas =[]
doencas = ["Peneumonia", "Insufciencia Cardaca", "Septicemia","AVC no Especificado", "Hemorragia Intracraniana"]

for x in range(6):
    c = df.iloc[:, x]
    colunas.append(c)

fig, ax = plt.subplots(figsize=(15, 8), dpi=200)

for i in range(5):
    ax.plot(colunas[0], colunas[i+1], label=doencas[i], marker='o')
    
media_anual = []

for i in range(len(colunas[0])):
    soma = 0
    for j in range(5):
        soma += colunas[j+1][i]
    m = soma/5
    media_anual.append(m)
    
ax.plot(colunas[0], media_anual, '--', color="#C71585", label="Media", marker='o')

plt.title('5 Maiores Causas de bitos em Hospitais no Brasil (2008 - 2018)', fontsize=17)
plt.xlabel('Ano', fontsize=17)
plt.ylabel('bitos', fontsize=17)
plt.ylim(0,60000)
plt.legend()
plt.margins(0)
plt.show()
```
